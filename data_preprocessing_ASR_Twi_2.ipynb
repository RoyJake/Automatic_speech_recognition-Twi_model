```python
import os
from pydub import AudioSegment
import shutil

def is_audio_playable(file_path):
    """Check if the audio file is playable."""
    try:
        # Attempt to load the audio file
        AudioSegment.from_file(file_path)
        return True
    except Exception as e:
        print(f"Error with file {file_path}: {e}")
        return False

def move_corrupted_files(source_folder, corrupted_folder):
    """
    Check all audio files in the source folder.
    Move corrupted or unplayable ones to the corrupted folder.
    """
    if not os.path.exists(corrupted_folder):
        os.makedirs(corrupted_folder)

    for file_name in os.listdir(source_folder):
        file_path = os.path.join(source_folder, file_name)

        if os.path.isfile(file_path):
            # Check if the file is an audio file and playable
            if not is_audio_playable(file_path):
                print(f"Moving corrupted file: {file_name}")
                shutil.move(file_path, os.path.join(corrupted_folder, file_name))


# Set your folders here
source_folder = "/content/drive/MyDrive/Twi_ASR/processed_audio"
corrupted_folder = "/content/drive/MyDrive/Twi_ASR/corrupted_audio"

move_corrupted_files(source_folder, corrupted_folder)
print("Processing complete.")
```
